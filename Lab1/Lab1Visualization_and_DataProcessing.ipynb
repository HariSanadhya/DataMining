## Business Understanding

### Analysis of Data Channels
Analyze the values of the Data Channel attributes to find if:
1. the maximum number of Data Channels defined for a URL if URL has more then one possible Data Channel.
* a Data Channel is defined for all the URL's in the dataset.

```python
data_channel_columns = ['data_channel_is_lifestyle', 'data_channel_is_entertainment', 'data_channel_is_bus',
                        'data_channel_is_socmed', 'data_channel_is_tech', 'data_channel_is_world']
df_news[data_channel_columns].apply(sum, axis=1).describe()
```

The above output shows that the maximum sum of the data channel values is 1 meaning that for any record, there is maximum one data channel defined. Since the 25 percentile is 1, this mean that majority of the records (more than 75 percent, mean value= 0.846092 indicates that 84.6% data) have the single data channel defined. Based on this, we are assuming that for any given URL, there can be only one possible data channel.
Also since the minimum value is zero, this mean that the data channel is not defined for some of the records (1-0.846092 = approx 15.4% records). There can be a few possibilities here:
1. Data channel is other than the 6 channels defined. 
* Issue while collecting the data from the data source.
* The URL was released without categorizing the data channel.

Since the logic of defining the data channel is not known, we decided to create a new data channel namely __'miscellaneous'__ for the URL's whose data channel is not defined.
Also since we now know that an URL can be associated to maximum one data channel, so instead of having 6 different attributes holding the value of the data channel, it makes more sense to create a single categorical variable containing the data channel value. 
Below few section of code illustrates what we have discussed here.

```python
# Create a new categorical attribute named data_channel having the data channel defined.
df_news['data_channel'] = df_news[data_channel_columns].apply(lambda x:  
                                        'lifestyle' if (int(x[0])==1) 
                                        else 'entertainment' if int(x[1])==1 
                                        else 'business' if int(x[2])==1
                                        else 'social media' if int(x[3])==1
                                        else 'technology' if int(x[4])==1
                                        else 'world' if int(x[5])==1
                                        else 'miscellaneous', axis=1
                                       ) 
```

### Analysis of weekday fields
Analyze values in all the 7 weekday fields to answers the questions below:
1. Are the URL's published on a day or there are multiple days populated
* Does all the URL's have published day populated 

```python
weekday_columns = ['weekday_is_monday', 'weekday_is_tuesday', 'weekday_is_wednesday', 'weekday_is_thursday',
                   'weekday_is_friday','weekday_is_saturday', 'weekday_is_sunday']
df_news[weekday_columns].apply(sum, axis=1).describe()
```

Since the minimum value and the maximum value from the above output (sum of weekday field value for every row in the dataset) is 1, this mean that there is one and only one value defined in the entire dataset for the weekday fields which is logical as well because the URL cannot be published on multiple days. Since for any record, only one of these attributes will be defined, its more logical to create a single categorical variable instead of multiple attributes. 
So we created an additional attribute namely __published_on_day_of_week__ having the following values
* __1:__  if published day of the week in monday
* __2:__  if published day of the week in tuesday
* __3:__  if published day of the week in wednesday
* __4:__  if published day of the week in thursday
* __5:__  if published day of the week in friday
* __6:__  if published day of the week in saturday
* __7:__  if published day of the week in sunday

```python
df_news['published_on_day_of_week'] =  df_news[weekday_columns].apply(lambda x:  
                                                    1 if (int(x[0])==1) 
                                                    else 2 if int(x[1])==1 
                                                    else 3 if int(x[2])==1
                                                    else 4 if int(x[3])==1
                                                    else 5 if int(x[4])==1
                                                    else 6 if int(x[5])==1
                                                    else 7 if int(x[6])==1
                                                    else None, axis=1
                                                   )
```

## Exceptional Work: Dimensionality Reduction - PCA

### Principal Component Analysis

The first step is to create a dataframe with only the appropriate attributes that will be used as a training dataset for the PCA analysis. Categorical variables cannot be used and missing values are not allowed.  We will create a train dataframe from df_news dataframe and remove the extra variables. 

```python
# Make the copy and document all the attributes to see which need to be removed or converted to
#    numeric

df_pca_train = df_news.copy()
df_pca_train.describe(include='all')
df_news.info()
```

Address the non-numeric, drop the data_channel, published_on_day_of_week,url and none log columns that were converted to logs.

We will be using the PCA results to predict # of shares and since its a response variable, we will remove that from the datafrom.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

df_pca_train.drop('data_channel',axis=1, inplace=True)
df_pca_train.drop('published_on_day_of_week',axis=1, inplace=True)
df_pca_train.drop('url',axis=1, inplace=True)
df_pca_train.drop('num_imgs',axis=1 , inplace=True)
df_pca_train.drop('num_videos',axis=1 , inplace=True)
df_pca_train.drop('n_tokens_content',axis=1 , inplace=True)
df_pca_train.drop('num_hrefs',axis=1 , inplace=True)
df_pca_train.drop('num_self_hrefs',axis=1 , inplace=True)
df_pca_train.drop('shares',axis=1 , inplace=True)

pca_train_scaled = scale(df_pca_train)
```

```python
pca = PCA(n_components=41)
pca.fit(pca_train_scaled)
pca_transform = pca.transform(pca_train_scaled)

var = pca.explained_variance_ratio_
var1 = np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)

print(var1)
```

```python
# Make the scree plot to determine the appropriate number of principal components

xmin = 0
xmax = 100

fig = plt.figure(figsize=(15,8)) 
fig.suptitle('PCA scree Plot', fontsize=20)
plt.xlabel('Component Number')
plt.ylabel('Cumulative Variance Explained')
plt.xlim(xmin,xmax)

plt.plot(var1)
```
