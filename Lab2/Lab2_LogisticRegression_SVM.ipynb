## Project and Data Overview
The dataset chosen for this lab assignment is different than the dataset used in lab 1. The reason being the response variable in the lab 1 dataset was continous but for this assignment, since categorical response was required, we switched the dataset.
This dataset is from UCI machine learning repository and is named as Adult dataset or Census Income dataset. The purpose of collection of this data is to predict whether individual income exceeds \$50000/yr based on census data. <br>

** Attribute Information **
* age – Continuous variable, list the age of the individual in years.
* workclass – Categorical variable, defines the type of employer the individual has. Whether they are government, military, private, an d so on. Possible categories listed in the dataset are Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay and Never-worked
* fnlwgt - Continuous variable, list the number of people the census takers believe that observation represents.
* education – Categorical variable, lists the highest level of education achieved for that individual. Recorded values are Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th and Preschool.
* education_num - Categorical variable, lists the highest level of education achieved for that individual in numerical form. Recorded values are within the range 1 to 16
* marital_status - Continuous variable, defines the marital status of the individual, recorded values are Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse
* occupation - Continuous variable, defines the occupation of the individual, recorded values are Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv and Armed-Forces
* relationship - Continuous variable, contains family relationship values like husband, father, and so on, but only contains one per observation. Recorded values are Wife, Own-child, Husband, Not-in-family, Other-relative and Unmarried
* race - Continuous variable, descriptions of the individuals race. Black, White, Eskimo, and so on. Recorded values are White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other and Black
* sex – Categorical variable, represents the Biological Sex of the individual. Recorded values are Female and Male
* capital_gain – Continuous variable, defines the recorded Capital gains for the individual. 
* capital_loss – Continuous variable, defines the recorded Capital losses for the individual
* hours_per_week – Continuous variable, number of hours the individual works in a week
* native_country – Categorical variable, represents the country of origin for the person. Recorded values are United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong and Holand-Netherlands

**fnlwgt variable handling**<br><br>
The variable fnlwgt represents number of people that the census takers believe that observation represents. So this means that this variable represents the count of number of times the same record was in the dataset. By means of this variable, the size of a very huge dataset was reduced to a small dataset.<br>
As this variable does not represent the characterstic of the individual, this mean that this variable cannot be included directly into the model. We can expand the dataset to the original value by repeating each record in the dataset by the value contained in this field. This will make the dataset very huge and that too, that too without giving much information. <br>
So we thought to change this field to percent field and then find the closest number to convert it into integer value and then once done, expand the dataset as per the value of this field.

```python
if 'fnlwgt' in df_salary_50k:
    tot_fnlwgt = df_salary_50k['fnlwgt'].sum()
    df_salary_50k['fnlwgt1'] = df_salary_50k['fnlwgt']/tot_fnlwgt * 1000000.00
    df_salary_50k['fnlwgt1'] = df_salary_50k['fnlwgt1'].apply(lambda x: int(round(x,0)))
df_salary_50k.fnlwgt1.describe()
```

```python
if 'fnlwgt' in df_salary_50k:
    tot_fnlwgt = df_salary_50k['fnlwgt'].sum()
    df_salary_50k['fnlwgt1'] = df_salary_50k['fnlwgt']/tot_fnlwgt * 500000.00
    df_salary_50k['fnlwgt1'] = df_salary_50k['fnlwgt1'].apply(lambda x: int(round(x,0)))
df_salary_50k.fnlwgt1.describe()
```

### Split Training and Test Data
Randomly split data into train and test in the ratio of 4:1 with the process repeated 5 times. For this, we created a Function that accepts 2 paramaters - one is a dataframe and another is a string which is the response variable name within the dataframe passed. The function perfroms the following steps:-
* Converts all categorical variables in the dataframe into dummy/indicator variables
* Perfrom the 80:20 split to the dataframe to form train and test dataframes (Stratified Shuffle split used and the is iterated 5 times) 
* Create the logistic regression model and compute the model parameters including accuracy, confusion matrix
* Of the 5 iterations ran, for the last iteration, plot the weights.

```python
# run logistic regression and vary some parameters
from sklearn.linear_model import LogisticRegression as lr
from sklearn import metrics as mt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit 
from sklearn.datasets import make_classification
from sklearn.svm import LinearSVC
```

```python
def func_lr(df_object,y_name):
    dummy_df_object = pd.get_dummies(df_object)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_object.making_GT_50K.values 
    X = dummy_df_object[dummy_df_object.columns.difference([y_name])].values

    #  Set up the training and testing split function
    num_cv_iterations = 5
    iter_num = 0
    num_instances = len(y)
    cv_object =  StratifiedShuffleSplit(n_splits=num_cv_iterations,
                             test_size  = 0.2)
    lr_obj = lr(solver='saga', multi_class ='multinomial', class_weight='')
    # okay, so run through the cross validation loop and set the training and testing variable for one single iteration
    for train_indices, test_indices in cv_object.split(X,y): 
        
        X_train = X[train_indices]
        y_train = y[train_indices]

        X_test = X[test_indices]
        y_test = y[test_indices]

        lr_obj = lr()
        lr_obj.fit(X_train,y_train)  # train object
        y_hat = lr_obj.predict(X_test) # get test set precitions
        iter_num += 1
        # now let's get the accuracy and confusion matrix for this iterations of training/testing
        acc = mt.accuracy_score(y_test,y_hat)
        conf = mt.confusion_matrix(y_test,y_hat)
        print("====Iteration",iter_num," ====")
        print("accuracy", acc )
        print("confusion matrix\n",conf)
    weights_logestic_regression = pd.Series(lr_obj.coef_[0], index=dummy_df_object.columns.difference([y_name]))
    plt.figure(figsize=(16,9))
    weights_logestic_regression.plot(kind='bar')
    plt.show()
```

### Issues in the above code
The accuracy of the model is ok but when we look at the weights, we can see certain issues. 
* The weight of hours_per_week is negative which means that with the increase in the number of work hours, there is less chance that the individual will be making above $50K
* Age is having positive weight but is having very small value which lead to a conclusion like with experience, the change in income is very small. This statement is not always true. 
* All the categories in Race and workclass have negative weigths. This is doesnt appear to be true as per the understanding of the real world scenarios.

The cause of these issues could be unscaled data and even overfitting. So lets scale the data before applying logistic regression

```python
def func_scaled_lr(df_object,y_name):
    dummy_df_object = pd.get_dummies(df_object)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_object.making_GT_50K.values 
    X = dummy_df_object[dummy_df_object.columns.difference([y_name])].values

    #  Set up the training and testing split function
    num_cv_iterations = 5
    iter_num = 0
    num_instances = len(y)
    cv_object =  StratifiedShuffleSplit(n_splits=num_cv_iterations,
                             test_size  = 0.2)
    lr_obj = lr(solver='saga', multi_class ='multinomial', class_weight='')
    # okay, so run through the cross validation loop and set the training and testing variable for one single iteration
    for train_indices, test_indices in cv_object.split(X,y): 
        
        X_train = X[train_indices]
        y_train = y[train_indices]
        
        X_test = X[test_indices]
        y_test = y[test_indices]

        std_scl = StandardScaler()
        std_scl.fit(X_train)
        X_train_scaled = std_scl.transform(X_train) # apply to training
        X_test_scaled = std_scl.transform(X_test) 

        lr_obj = lr()
        lr_obj.fit(X_train_scaled,y_train)  # train object
        y_hat = lr_obj.predict(X_test_scaled) # get test set precitions
        iter_num += 1
        # now let's get the accuracy and confusion matrix for this iterations of training/testing
        acc = mt.accuracy_score(y_test,y_hat)
        conf = mt.confusion_matrix(y_test,y_hat)
        print("====Iteration",iter_num," ====")
        print("accuracy", acc )
        print("confusion matrix\n",conf)
    weights_logestic_regression = pd.Series(lr_obj.coef_[0], index=dummy_df_object.columns.difference([y_name]))
    plt.figure(figsize=(16,9))
    weights_logestic_regression.plot(kind='bar')
    plt.show()
```

#### Stochastic Gradient Descent

```python
%%time
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit 
from sklearn.model_selection import ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn import metrics as mt
import matplotlib.pyplot as plt
%matplotlib inline

def func_grad_descent(df_input, y_variable):
    dummy_df_salary_50k = pd.get_dummies(df_input)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_salary_50k[str(y_variable)].values 
    X = dummy_df_salary_50k[dummy_df_salary_50k.columns.difference([str(y_variable)])].values

    #  Set up the training and testing split function
    num_cv_iterations = 5
    num_instances = len(y)
    cv_object = StratifiedShuffleSplit(n_splits=num_cv_iterations,test_size  = 0.2)
    regularize_const = 0.1
    svm_sgd = SGDClassifier(alpha=regularize_const, fit_intercept=True, l1_ratio=0.0, learning_rate='optimal',
                            loss='hinge', n_iter=num_cv_iterations, n_jobs=-1, penalty='l2')

    # X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
    for train_indices, test_indices in cv_object.split(X,y): 
        X_train = X[train_indices]
        y_train = y[train_indices]    
        X_test = X[test_indices]
        y_test = y[test_indices]

        std_scl = StandardScaler()
        std_scl.fit(X_train)
        X_train_scaled = std_scl.transform(X_train) # apply to training
        X_test_scaled = std_scl.transform(X_test) 

    #     scl = StandardScaler()
    #     for train_idx, test_idx in cv_object.split(X,y):
        svm_sgd.fit(X_train_scaled,y[train_indices]) # stochastic
        #svm_sgd.partial_fit(scl.fit_transform(X[train_idx]),y[train_idx], classes = np.unique(y_train)) #minibatch
        yhat = svm_sgd.predict(X_test_scaled)

        conf = mt.confusion_matrix(y_test,yhat)
        acc = mt.accuracy_score(y_test,yhat)

        print('***Iterating****************')
        print('Accuracy:', acc )
        print('Confusion Matrix:')
        print(conf)
        # look at the support vectors
        # print('Support Vector Shape:', svm_sgd.support_vectors_.shape)
        #print('Support Shape:', svm_sgd.)
    weights = pd.Series(svm_sgd.coef_[0],index=dummy_df_salary_50k.columns.difference([str(y_variable)]).values)
    plt.figure(figsize = (40,25))     
    weights.plot(kind='bar')
    plt.tick_params(axis='x', labelsize=30)
    plt.tick_params(axis='y', labelsize=30)
```

#### MiniBatch Gradient Descent

```python
def func_minibatch_grad_descent(df_input, y_variable):
    dummy_df_salary_50k = pd.get_dummies(df_input)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_salary_50k[str(y_variable)].values 
    X = dummy_df_salary_50k[dummy_df_salary_50k.columns.difference([str(y_variable)])].values

    #  Set up the training and testing split function
    num_cv_iterations = 5
    num_instances = len(y)
    cv_object = StratifiedShuffleSplit(n_splits=num_cv_iterations,test_size  = 0.2)
    regularize_const = 0.1
    svm_sgd = SGDClassifier(alpha=regularize_const, fit_intercept=True, l1_ratio=0.0, learning_rate='optimal',
                            loss='hinge', n_iter=num_cv_iterations, n_jobs=-1, penalty='l2')

    # X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
    for train_indices, test_indices in cv_object.split(X,y): 
        X_train = X[train_indices]
        y_train = y[train_indices]    
        X_test = X[test_indices]
        y_test = y[test_indices]

        std_scl = StandardScaler()
        std_scl.fit(X_train)
        X_train_scaled = std_scl.transform(X_train) # apply to training
        X_test_scaled = std_scl.transform(X_test) 

        svm_sgd.partial_fit(X_train_scaled,y_train, classes = np.unique(y_train)) #minibatch
        yhat = svm_sgd.predict(X_test_scaled)

        conf = mt.confusion_matrix(y_test,yhat)
        acc = mt.accuracy_score(y_test,yhat)

        print('***Iterating****************')
        print('Accuracy:', acc )
        print('Confusion Matrix:')
        print(conf)
        # look at the support vectors
        # print('Support Vector Shape:', svm_sgd.support_vectors_.shape)
        #print('Support Shape:', svm_sgd.)
    weights = pd.Series(svm_sgd.coef_[0],index=dummy_df_salary_50k.columns.difference([str(y_variable)]).values)
    plt.figure(figsize = (40,25))     
    weights.plot(kind='bar')
    plt.tick_params(axis='x', labelsize=30)
    plt.tick_params(axis='y', labelsize=30)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit 
from sklearn.model_selection import ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics as mt
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
%matplotlib inline


# Dataset proved to be too large to execute in a reasonable amount of time using SVM Model.  Here we are subsampling
new_df_salary_50k = modified_df_salary_50k.sample(frac=.05, replace=True)
dummy_df_salary_50k = pd.get_dummies(new_df_salary_50k)

# Utility function to move the midpoint of a colormap to be around
# the values of interest.
class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))

# pipeline function using grid search to automatically find best C and gamma parameters
def get_best_cost_gamma(df, param):
    dummy_df_salary_50k = pd.get_dummies(df)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_salary_50k.making_GT_50K.values 
    X = dummy_df_salary_50k[dummy_df_salary_50k.columns.difference(['making_GT_50K'])].values
    
    #  Set up the training and testing split function
    num_cv_iterations = 3
    num_instances = len(y)
    cv_object = StratifiedShuffleSplit(n_splits=num_cv_iterations,test_size  = 0.2)

    # X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
    for train_indices, test_indices in cv_object.split(X,y): 
        X_train = X[train_indices]
        y_train = y[train_indices]    
        X_test = X[test_indices]
        y_test = y[test_indices]
        # scale
        std_scl = StandardScaler()
        std_scl.fit(X_train)
        X_train_scaled = std_scl.transform(X_train) # apply to training
        X_test_scaled = std_scl.transform(X_test) 
        
        # start pipeline
        pipeline = Pipeline([('clf', SVC(kernel='rbf', gamma=0.01, C=100))])
        print(X_train.shape)
        grid_search = GridSearchCV(pipeline, param, n_jobs=2, verbose=1, scoring='accuracy')
        grid_search.fit(X_train_scaled, y_train)
        print('***Iterating*************')
        print('Best score: %0.3f' % grid_search.best_score_)
        print('Best parameters set:')
        best_parameters = grid_search.best_estimator_.get_params()
        for param_name in sorted(param.keys()):
            print('\t%s: %r' % (param_name, best_parameters[param_name]))
        predictions = grid_search.predict(X_test_scaled)
        print(classification_report(y_test, predictions))
    
     # plot heatmap of gamma versus C 
    C_range = param['clf__C']
    gamma_range = param['clf__gamma']
    scores = grid_search.cv_results_['mean_test_score'].reshape(len(C_range),
                                                         len(gamma_range))
    plt.figure(figsize=(10, 12))
    plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
    plt.imshow(scores, interpolation='nearest', cmap=plt.cm.hot,
               norm=MidpointNormalize(vmin=0.2, midpoint=0.92))
    plt.xlabel('gamma')
    plt.ylabel('C')
    plt.colorbar()
    plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
    plt.yticks(np.arange(len(C_range)), C_range)
    plt.title('Validation accuracy')
    plt.show() 
    

# run svm model with given hyperparameters
# xtrs = X_train_scaled, xtss = x_test_scaled, y_trn = y_train, y_tst = y_test, trn_idx = train_indices 
def run_svm(xtrs, xtss, y_trn, y_tst, trn_idx, kde, ker ='rbf', c = 30, gam = 0.01):
    svm_clf = SVC(C = c, kernel= ker, degree=3, gamma = gam) # get object
    svm_clf.fit(xtrs, y_trn)  # train object
    y_hat = svm_clf.predict(xtss) # get test set predictions

    acc = mt.accuracy_score(y_tst,y_hat)
    conf = mt.confusion_matrix(y_tst,y_hat)
    print('***Iterating****************')
    print('Accuracy:', acc )
    print('Confusion Matrix:')
    print(conf)
    # look at the support vectors
    print('Support Vector Shape:', svm_clf.support_vectors_.shape)
    print('Support Shape:', svm_clf.support_.shape)
    print('Number of Support Vectors:', svm_clf.n_support_ )
    
    # Density plots if kde is true
    if kde:
        df_tested_on = dummy_df_salary_50k.iloc[trn_idx] # saved from above, the indices chosen for training
        # now get the support vectors from the trained model
        df_support = df_tested_on.iloc[svm_clf.support_,:]


        # group the original data and the support vectors
        df_grouped_support = df_support.groupby(['making_GT_50K'])
        df_grouped = new_df_salary_50k.groupby(['making_GT_50K'])

        # plot KDE of Different variables
        vars_to_plot = ['age','hours_per_week','capital_total','Have Dependents', 'sex']

        for v in vars_to_plot:
            plt.figure(figsize=(10,4))
            # plot support vector stats
            plt.subplot(1,2,1)
            ax = df_grouped_support[v].plot.kde() 
            plt.legend(['GT50K','LE50K'])
            plt.title(v+' (Instances chosen as Support Vectors)')

            # plot original distributions
            plt.subplot(1,2,2)
            ax = df_grouped[v].plot.kde() 
            plt.legend(['GT50K','LE50K'])
            plt.title(v+' (Original)')
    return svm_clf

# cross validate specifying hyperparams gamma, C and Kernel
def run_svm_cv(df, y_name, kde = False, gamma = .1, C = 1, ker = 'rbf'):
    dummy_df_salary_50k = pd.get_dummies(df)
    # Separate Explanatory (X) and reference (y) variables as matrices:
    y = dummy_df_salary_50k.making_GT_50K.values 
    X = dummy_df_salary_50k[dummy_df_salary_50k.columns.difference(['making_GT_50K'])].values
    
    #  Set up the training and testing split function
    num_cv_iterations = 3
    num_instances = len(y)
    cv_object = StratifiedShuffleSplit(n_splits=num_cv_iterations,test_size  = 0.2)

    # X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
    
    svm_ = []
    for train_indices, test_indices in cv_object.split(X,y): 
        X_train = X[train_indices]
        y_train = y[train_indices]    
        X_test = X[test_indices]
        y_test = y[test_indices]
        # scale
        std_scl = StandardScaler()
        std_scl.fit(X_train)
        X_train_scaled = std_scl.transform(X_train) # apply to training
        X_test_scaled = std_scl.transform(X_test) 

        # call svm function
        svm_ = run_svm(X_train_scaled, X_test_scaled, y_train, y_test, train_indices, kde, ker, C, gamma)
        
        if ker == 'linear':
            weights = pd.Series(svm_.coef_[0],index=dummy_df_salary_50k.columns.difference(['making_GT_50K']).values)
            plt.figure(figsize = (28,10))
            weights.plot(kind='bar')
```

#### Influence of Hyperparameters on SVC
* A high C value seems to try to classify all classes correctly by giving the model more ability to select more Support Vectors. A low C value runs faster as the decision plane is smoother thus less computationally intensive. C stands for 'cost' of mis-classifying data so with a high value of C SVC tries to avoid mis-classifying data by using more support vectors.
* A small gamma value tries to have deeper influence of the training instance and thus takes longer to process while a larger gamma value will process faster because it has a more shallow influence.
* Kernel types - 'rbf' is a Gaussian Model (normal distribution assumed at each data point) and stands for Radial Basis Function. It is ideal for complex data features and uses normal curves around data points. C and gamma influence the classification characteristic of this model. Linear uses linear regression for classification and 'poly' uses a similar approach with polynomials. The 'sigmoid' kernel is type of kernel that comes from Neural Networks field. Key parameters when using sigmoid are the alpha slope and the intercept constant c. Sigmoid and Poly were not implemented in this mini lab.
